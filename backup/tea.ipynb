```python
import pandas as pd

df = pd.read_csv('cofee_shop_sales.csv')

df.head()

filter_tea = df[df['product_category'].str.contains('Tea', case=False, na=False)]

# Asumsi df adalah dataframe Anda
result = filter_tea.groupby(['product_id','product_category','product_type','product_detail',]).agg({
    'transaction_qty': 'sum',
    'unit_price': 'mean' # Contoh agregasi lain
}).reset_index()

print(result)

result.to_csv('product_id_details.csv', index=False)
```
