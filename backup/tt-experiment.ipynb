```python
import pandas as pd
```

```python
df = pd.read_csv('cofee_shop_sales.csv')

df.head()
df.info()
```

```python
# Menghitung jumlah lalu mengubahnya menjadi DataFrame yang rapi
df_hasil = df.groupby('product_category').size().reset_index(name='jumlah')

# Menyimpan hasil ke file CSV
df_hasil.to_csv('jumlah_product_type.csv', index=False)
```
